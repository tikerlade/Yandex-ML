## Using of kNN

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import *
from sklearn.preprocessing import scale
```

```python
data = pd.read_csv('wine_data.csv', header=None)
data.head()
```

```python
X = pd.read_csv('wine_data.csv', header=None, usecols=list(range(1,14)))
y = pd.read_csv('wine_data.csv', header=None, usecols=[0]).values.reshape(len(X),)
X = scale(X)
#X = pd.DataFrame(scale(data.loc[:,2:]))
#X = data.loc[:,2:]
#y = data.loc[:,0]

#X.head()
#y.head()
```

```python
folder = KFold(n_splits=5, shuffle=True, random_state=42)
acc = []

for i in range(1, 51):
    classifier = KNeighborsClassifier(n_neighbors=i)
    classifier = classifier.fit(X, y)
    res = cross_val_score(classifier, X, y, cv=folder)
    
    print(res.mean())
    acc.append(res.mean())
print(len(acc))
print(round(max(acc), 2), acc.index(max(acc))+1)
```
