## Minkowski

```python
import pandas as pd
import numpy as np
from tqdm import tqdm_notebook
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import scale
from sklearn.datasets import load_boston
```

```python
data = load_boston()
names = data['feature_names']

X = pd.DataFrame(scale(data['data']), columns=names)
y = pd.DataFrame(data['target'], columns=['target'])
```

```python
folder = KFold(n_splits=5, shuffle=True, random_state=42)
out = list()


for parametr in tqdm_notebook(np.linspace(1,10,200)):
    regressor = KNeighborsRegressor(n_neighbors=5, weights='distance', p=parametr, metric='minkowski')
    res = cross_val_score(regressor, X, y, cv=folder)
    out.append(res.mean())


#print(out)
print(round(max(out),2), out.index(max(out))+1)
```
